```python
#import libraries
import requests
import lxml.html as lh
import pandas as pd
```

## *Step 1: For 2024, iterate through basketball-reference for per-game data.*

```python
url='https://www.basketball-reference.com/leagues/NBA_2024_per_game.html'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')

#######################################################
#Create empty list
col = []
i = 0
#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i += 1
    name = t.text_content()
    col.append((name,[]))

#######################################################
#Since the first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T = tr_elements[j]
    #If row is not of size 10, the //tr data is not from our table
    if len(T) != 30:
        break
    #i is the index of our column
    i = 0
    #Iterate through each element of the row
    for t in T.iterchildren():
        data = t.text_content()
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data = int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i += 1

#######################################################
#Create the DataTrame
Dict = {title:column for (title,column) in col}
data_df = pd.DataFrame(Dict)


############################OUTPUT THE DATA INTO EXCEL FILE###########################
with pd.ExcelWriter("/content/drive/MyDrive/NBA_project/NBA_Project/data/Data_PerGame.xlsx") as writer:
    data_df.to_excel(writer, index = False)
```
